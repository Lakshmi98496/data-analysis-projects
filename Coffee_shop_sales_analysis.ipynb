##☕ Coffee Shop Sales Data Analysis

In this project, we analyze a real-world coffee shop sales dataset stored in an Excel (.xlsx) format. The dataset contains information such as:

Transaction date and time

Payment type (cash/card)

Unique customer card ID

Amount spent

Type of coffee purchased


Our goal is to extract insights like the most popular coffee types, peak sales dates, preferred payment methods, and revenue trends — using Pandas, Seaborn, and Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel("Coffee.xlsx")
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm')
#Since there are no multiple numerical values we get heatmap for only one column
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel("Coffee.xlsx")
df.groupby('coffee_name')['money'].sum().plot(kind='bar')
```

```python
df.groupby('date')['money'].sum().plot(kind='line')
```
