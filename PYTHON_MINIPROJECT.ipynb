```python
#Day which has highest average bill
import pandas as pd
df=pd.read_csv("Tips.csv (1).csv",index_col=0)
df.groupby('Day')['TotalBill'].mean()
```

```python
#To find whether people who dine at night give more tips than those at lunch
import pandas as pd
df=pd.read_csv("Tips.csv (1).csv",index_col=0)
df.groupby('Time')['Tips'].mean()
```

```python
#To visualize the relationship between totalbill and tips
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Tips.csv (1).csv",index_col=0)
sns.scatterplot(x='TotalBill',y='Tips',data=df)
plt.title('Total Bill vs Tips')
plt.show()
```

```python
#Do people who smoke give higher tips on average?
import pandas as pd
df=pd.read_csv("Tips.csv (1).csv",index_col=0)
df.groupby('Smoker')['Tips'].mean()
```

```python
#Distribution of bill amounts
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(df['TotalBill'],bins=10)
plt.title('Distribution of Bill Amounts')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Tips.csv (1).csv",index_col=0)
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title('Distribution of Bill Amounts by Day')
plt.show()
```

```python
#To find top 5 custiomers who gave highest tips
import pandas as pd
df=pd.read_csv("Tips.csv (1).csv",index_col=0)
df.sort_values(by='Tips',ascending=False).head(5)
```

```python
#To show how tip amounts vary across days using boxplots
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(x='Day',y='TotalBill',data=df)
plt.title("Distribution of Bill Amounts by Day")
plt.show()
```
